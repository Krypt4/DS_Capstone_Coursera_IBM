```python
import pandas as pd

df = pd.read_csv("../data/dataset_part_2.csv")

df["PayloadMass"] = pd.to_numeric(df["PayloadMass"], errors="coerce")

max_payload = df["PayloadMass"].max()

max_boosters = df[df["PayloadMass"] == max_payload]["BoosterVersion"].unique()

print("Max payload mass:", max_payload)
print("Boosters:", max_boosters)
```
