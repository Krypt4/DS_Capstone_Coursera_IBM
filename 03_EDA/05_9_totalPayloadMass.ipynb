```python
import pandas as pd

df = pd.read_csv("../data/dataset_part_2.csv")

nasa_df = df[df["BoosterVersion"].str.contains("NASA", na=False)]

total_payload_nasa = nasa_df["PayloadMass"].sum()

print("Total payload NASA carried:", total_payload_nasa)

print("----------------")

spacex_df = df[df["BoosterVersion"].notna()]

total_payload = spacex_df["PayloadMass"].sum()

print("Total payload carried:", total_payload)
```
