```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/dataset_part_2.csv")

# 2. Crear el catplot para Payload vs LaunchSite
# Usamos 'aspect' para que el gráfico sea más ancho y fácil de leer
sns.catplot(
    data=df,
    x="PayloadMass",
    y="LaunchSite",
    hue="Class",
    aspect=2,
    height=5,
    s=10  # Tamaño de los puntos
)

# 3. Configuración estética
plt.title("Relación entre Carga Útil y Sitio de Lanzamiento por Resultado", fontsize=15)
plt.xlabel("Masa de Carga Útil (kg)", fontsize=12)
plt.ylabel("Sitio de Lanzamiento", fontsize=12)

plt.show()
```
