```python
import sqlite3
import pandas as pd

df = pd.read_csv("../data/dataset_part_2.csv")

conn = sqlite3.connect(":memory:")
df.to_sql("spacex", conn, index=False)

# Total éxitos
query1 = """
SELECT COUNT(*) as total_success
FROM spacex
WHERE Class = 1
"""
print(pd.read_sql(query1, conn))

# Promedio payload
query2 = """
SELECT AVG(PayloadMass) as avg_payload
FROM spacex
"""
print(pd.read_sql(query2, conn))
```
