```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset_part_2.csv")

df["PayloadMass"] = pd.to_numeric(df["PayloadMass"], errors="coerce")

df["Orbit"] = df["Orbit"].fillna("Unknown")

df_clean = df.dropna(subset=["PayloadMass"])

print("Filas válidas:", df_clean.shape[0])

plt.figure(figsize=(10,5))

sns.scatterplot(
    data=df_clean,
    x="PayloadMass",
    y="Orbit",
    hue="Class"
)

plt.title("Payload vs Orbit Type")
plt.show()
```
