```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset_part_2.csv")

df["PayloadMass"] = pd.to_numeric(df["PayloadMass"], errors="coerce")

df_clean = df[df["PayloadMass"].notna()].copy()

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df_clean,
    x="PayloadMass",
    y="LaunchSite",
    hue="Class"
)

plt.title("Payload vs Launch Site")
plt.show()
```
