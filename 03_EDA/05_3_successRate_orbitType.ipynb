```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset_part_2.csv")

df = df.dropna(subset=["Orbit"])

df["Orbit"] = df["Orbit"].str.strip()

orbit_success = df.groupby("Orbit", as_index=False)["Class"].mean()

plt.figure(figsize=(12,5))

sns.scatterplot(
    data=orbit_success,
    x="Orbit",
    y="Class"
)

plt.xticks(rotation=90)
plt.ylim(0, 1)
plt.title("Success Rate vs Orbit Type")
plt.show()
```
