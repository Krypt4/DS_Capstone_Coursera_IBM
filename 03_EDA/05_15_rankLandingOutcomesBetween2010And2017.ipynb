```python
import pandas as pd

df = pd.read_csv("../data/dataset_part_2.csv")

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

df_range = df[
    (df["Date"] >= "2010-06-04") &
    (df["Date"] <= "2017-03-20")
]

outcome_counts = df_range["Outcome"].value_counts().reset_index()

outcome_counts.columns = ["Outcome", "Count"]

outcome_counts = outcome_counts.sort_values("Count", ascending=False)

print(outcome_counts)
```
