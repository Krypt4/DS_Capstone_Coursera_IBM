```python
import pandas as pd

df = pd.read_csv("../data/dataset_part_1.csv")

# Crea columna Class
def classify(outcome):
    if outcome == True:
        return 1
    else:
        return 0

df["Class"] = df["Outcome"].apply(classify)

# Limpieza básica
df = df.dropna(subset=["FlightNumber"])

df.to_csv("../data/dataset_part_2.csv", index=False)

df.head()
```
