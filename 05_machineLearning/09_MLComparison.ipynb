```python
import matplotlib.pyplot as plt

models = {'Logistic Regression': 0.846, 'SVM': 0.848, 'Decision Tree': 0.875, 'KNN': 0.846}

plt.figure(figsize=(10, 6))
plt.bar(models.keys(), models.values(), color=['blue', 'green', 'orange', 'red'])

plt.xlabel('Classification Models', fontsize=12)
plt.ylabel('Accuracy Score', fontsize=12)
plt.title('Build Model Accuracy Comparison', fontsize=15)
plt.ylim(0.8, 0.9)
plt.show()

best_model = max(models, key=models.get)
print(f"The model with the highest accuracy is: {best_model}")
```
